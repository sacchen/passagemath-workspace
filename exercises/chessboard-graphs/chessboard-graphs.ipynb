# Chessboard Graphs in passagemath

**What this notebook teaches:** how to construct chessboard graphs using `sage.graphs.generators.chessboard`, and what the `dim_list` parameter means.

**Function explored:** `QueenGraph` — vertices are squares of a board, edges connect squares a queen can reach in one move.

**Structure:** setup → construct → inspect → exercise. Run cells in order.

```python
# Setup — run this first.
from sage.graphs.generators.chessboard import QueenGraph
print("OK")
```

---
## Section 1: Construct a Queen Graph

`QueenGraph(dim_list)` takes a **list** of board dimensions, not two integers.

- `[3, 3]` means a 3×3 board
- Each vertex is a square; each edge means a queen can move between those two squares in one step

```python
G = QueenGraph([3, 3])
print(G)
print(f"Vertices: {G.order()}")
print(f"Edges:    {G.size()}")
```

---
## Section 2: Inspect the graph

Vertices are tuples `(row, col)`. Edges connect squares a queen can reach.

```python
print("Vertices:", sorted(G.vertices()))
print()
print("Edges from (0,0):")
for neighbor in sorted(G.neighbors((0, 0))):
    print(" ", (0, 0), "->", neighbor)
```
