# Chessboard Graphs in passagemath

**What this notebook teaches:** how to construct chessboard graphs using `sage.graphs.generators.chessboard`, and what the `dim_list` parameter means.

**Function explored:** `QueenGraph` — vertices are squares of a board, edges connect squares a queen can reach in one move.

**Structure:** setup → construct → inspect → exercise. Run cells in order.

```python
# Setup — run this first.
from sage.graphs.generators.chessboard import QueenGraph

# Fix plot rendering in plain Python Jupyter kernels.
# Workaround for passagemath/passagemath#2236, fixed in #2351 but not yet released.
from io import BytesIO
from matplotlib import rcParams
from matplotlib.backends.backend_agg import FigureCanvasAgg
from sage.plot.graphics import Graphics

def _graphics_repr_png(self):
    options = {}
    options.update(self.SHOW_OPTIONS)
    options.update(self._extra_kwds)
    dpi = options.pop('dpi')
    transparent = options.pop('transparent')
    fig_tight = options.pop('fig_tight')
    rc_backup = (rcParams['ps.useafm'], rcParams['pdf.use14corefonts'],
                 rcParams['text.usetex'])
    try:
        figure = self.matplotlib(**options)
        figure.set_canvas(FigureCanvasAgg(figure))
        figure.tight_layout()
        opts = {'dpi': dpi, 'transparent': transparent}
        if fig_tight:
            opts['bbox_inches'] = 'tight'
        extra = getattr(self, '_bbox_extra_artists', None)
        if extra:
            opts['bbox_extra_artists'] = extra
        buf = BytesIO()
        figure.savefig(buf, format='png', **opts)
        return buf.getvalue()
    finally:
        (rcParams['ps.useafm'], rcParams['pdf.use14corefonts'],
         rcParams['text.usetex']) = rc_backup

Graphics._repr_png_ = _graphics_repr_png

print("OK")
```

---
## Section 1: Construct a Queen Graph

`QueenGraph(dim_list)` takes a **list** of board dimensions, not two integers.

- `[3, 3]` means a 3×3 board
- Each vertex is a square; each edge means a queen can move between those two squares in one step

```python
G = QueenGraph([3, 3])
print(G)
print(f"Vertices: {G.order()}")
print(f"Edges:    {G.size()}")
```

```python
G.plot()
```

---
## Section 2: Inspect the graph

Vertices are tuples `(row, col)`. Edges connect squares a queen can reach.

```python
print("Vertices:", sorted(G.vertices()))
print()
print("Edges from (0,0):")
for neighbor in sorted(G.neighbors((0, 0))):
    print(" ", (0, 0), "->", neighbor)
```
